```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
metar_data = pd.read_csv('../00_data_preparation/metar/metar_dataset_with_airport_location.csv')
```

```python
metar_data
```

```python
metar_fz = metar_data[metar_data['contain_FZ']]
```

```python
metar_fz
```

```python
metar_fz['issued_at'] = pd.to_datetime(metar_fz['issued_at'])

metar_fz['month'] = metar_fz['issued_at'].dt.month

month_count = metar_fz.groupby('month').size()


# FZ
plt.figure(figsize=(10, 6))
sns.barplot(x=month_count.index, y=month_count.values, palette='viridis')
plt.title('FZ occurrence in METAR per month')
plt.xlabel('Month')
plt.ylabel('Number of METAR with FZ')
plt.show()

# Metartype
plt.figure(figsize=(10, 6))
sns.countplot(x='month', hue='metartype', data=metar_fz, palette='viridis', dodge=True)
plt.title('Number of METAR with FZ per metartype')
plt.xlabel('Month')
plt.ylabel('Number of METAR with FZ')
plt.show()
```
